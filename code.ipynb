```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=0)

accuracy_scores={}
```

```python
# KNN
from sklearn import neighbors
from sklearn.metrics import accuracy_score

clf_knn=neighbors.KNeighborsClassifier(n_neighbors=5)
clf_knn.fit(X_train,y_train)
pred_knn=clf_knn.predict(X_test)
acc_knn = accuracy_score(y_test,pred_knn)

accuracy_scores['KNN Classifier']=acc_knn

print("Accuracy using KNN Classifier: ", acc_knn)     
```

```python
# decision tree classifier
from sklearn import tree

criteria=['gini', 'entropy']
splitters=['best', 'random']
acc_score=[]
for criterion in criteria:
    for splitter in splitters:
        for i in range (1, 20):
            clf_tree=tree.DecisionTreeClassifier(criterion=criterion, splitter=splitter, max_depth=i, random_state=0)
            clf_tree.fit(X_train,y_train)
            pred_tree=clf_tree.predict(X_test)
            acc = accuracy_score(y_test,pred_tree)
            acc_score.append(acc)
            acc_tree=max(acc_score)

accuracy_scores['Decision Tree Classifier']=acc_tree

print("Accuracy using Decision Tree Classifier:", acc_tree)
```

```python
# Gradient Boosted Decision Tree
from sklearn.ensemble import GradientBoostingClassifier

clf_gb = GradientBoostingClassifier(random_state=0).fit(X_train, y_train)
pred_gb=clf_gb.predict(X_test)
acc_gb=accuracy_score(y_test, pred_gb)

accuracy_scores['Gradient Boosted Decision Tree Classifier']=acc_gb

print("Accuracy using Gradient Boosted Decision Tree Classifier:", acc_gb)
```

```python
# Linear SVC
from sklearn.svm import LinearSVC

# kernels=['linear', 'poly', 'rbf', 'sigmoid']
# gammas=['scale', 'auto']
# acc_score=[]

clf_svc= LinearSVC(multi_class='crammer_singer', C=10, random_state=45).fit(X_train, y_train)
pred_svc=clf_svc.predict(X_test)
acc_svc=accuracy_score(y_test, pred_svc)
# acc_score.append(acc)
# acc_svc=max(acc_score)

accuracy_scores['Linear SVC']=acc_svc

print("Accuracy using SVC:", acc_svc)
```

```python
# Kernelized SVC
from sklearn.svm import SVC

kernels=['poly', 'linear', 'rbf']
acc_score=[] 
for kernel in kernels:
    for i in range (1, 20, 3):
        for j in range (1, 20, 3):
            clf_ksvm = SVC(kernel = kernel, degree = i, gamma=j, decision_function_shape='ovr',C=10).fit(X_train, y_train)
            pred_ksvm=clf_ksvm.predict(X_test)
            acc=accuracy_score(y_test, pred_ksvm)
            acc_score.append(acc) 
            acc_ksvc=max(acc_score)

accuracy_scores['Kernelized SVC']=acc_ksvc

print("Accuracy using Kernelized SVC:", acc_ksvc)
```

```python
# Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB

clf_nb=GaussianNB().fit(X_train, y_train)
pred_nb=clf_nb.predict(X_test)
acc_nb=accuracy_score(y_test, pred_nb)

accuracy_scores['Naive Bayes Classifier']=acc_nb

print("Accuracy using Naive Bayes Classifier:", acc_nb)
```

```python
# Random Forests Classification
from sklearn.ensemble import RandomForestClassifier

criteria=['gini', 'entropy']
acc_score=[]  
for criterion in criteria:
    for i in range(1, 20):
        clf_rf = RandomForestClassifier(criterion=criterion, max_depth=i, bootstrap=False, random_state=0).fit(X_train, y_train)
        pred_rf=clf_rf.predict(X_test)
        acc=accuracy_score(y_test, pred_rf)
        acc_score.append(acc)
        acc_rf=max(acc_score)

accuracy_scores['Random Forest Classifier']=acc_rf

print("Accuracy using Random Forest Classifier:", acc_rf)
```
